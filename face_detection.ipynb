# Detect faces of images

```python
from google.colab.patches import cv2_imshow
import cv2
import os
```

```python
face_detector = cv2.CascadeClassifier('/content/drive/Othercomputers/My Laptop/Documents/Deep_learning/CNN/face detection/opencv/data/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def getFaces(img_path):
    img = cv2.imread(img_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_detector.detectMultiScale(img_gray, 1.3, 5)
    count = 0
    for x, y, w, h in faces:
        img_face = cv2.resize(img[y+3: y+h-3, x+3: x+w-3], (64, 64))
        cv2.imwrite(img_path.replace('img', 'face_cut').split('.jpg')[0] + f'_{count}.jpg', img_face)
        # cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        count += 1
```

```python
folder_path = '/content/drive/Othercomputers/My Laptop/Documents/Deep_learning/CNN/face detection/'
# image_path = f'{folder_path}img1.jpg'
```

```python
for whatelse in os.listdir(f'{folder_path}img/'):
    # print(whatelse)
    whatelse_path = os.path.join(f'{folder_path}img/', whatelse)
    for sub_whatelse in os.listdir(whatelse_path):
        # print(sub_whatelse)
        img_path = os.path.join(whatelse_path, sub_whatelse)
        if not os.path.isdir(whatelse_path.replace('img', 'face_cut')):
            os.mkdir(whatelse_path.replace('img', 'face_cut'))
        if img_path.endswith('.jpg'):
            # print(img_path)
            getFaces(img_path)

cv2.destroyAllWindows()
```
